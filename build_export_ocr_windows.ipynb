```python
# delete build files and dist files
import os
import shutil

def delete_build_dist():
    if os.path.exists('build'):
        shutil.rmtree('build')
    if os.path.exists('dist'):
        shutil.rmtree('dist')

delete_build_dist()
```

```python

# Specify the source and destination paths
exe_name = 'ocr_server.exe'

source_path = './dist/' + exe_name

dest_folder_name = '~/Documents/GitHub/APOS/ocr_server/'
destination_dir = os.path.expanduser(dest_folder_name)
```

```python
# check if the destination directory exists, else create it
if not os.path.exists(destination_dir):
    os.makedirs(destination_dir)

# delete all files in the destination directory
for filename in os.listdir(destination_dir):
    file_path = os.path.join(destination_dir, filename)
    try:
        if os.path.isfile(file_path) or os.path.islink(file_path):
            os.unlink(file_path)
        elif os.path.isdir(file_path):
            shutil.rmtree(file_path)
    except Exception as e:
        print('Failed to delete %s. Reason: %s' % (file_path, e))

# Copy the file
destination_path = os.path.join(destination_dir, exe_name)
shutil.copyfile(source_path, destination_path)
```
